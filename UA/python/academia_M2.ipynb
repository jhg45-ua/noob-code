```python
# Graficar funciones
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100) # Rango en X | Entre x=-5 y x=5 en 100 intervalos
y = np.sin(x)
plt.plot(x,y,color='red')
plt.title('sin(x)')
plt.plot(0,0,'ro' ,color='blue', alpha=0.3)
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.grid(color='green', alpha=0.3)
plt.show()
```

```python
# Resolver una ecuacion
from sympy import *
x = Symbol('x')

f = x**2-4*x
df = diff(f,x)
ecuacion1 = Eq(f, 0)
ecuacion2 = Eq(df,0)
soluciones = solve(ecuacion1, x)

print('Puntos de corte con el eje x de la funcion', f, 'son', soluciones, 'y su derivada es', df)
```

```python
from sympy import solve, Eq
x, y, z = symbols('x,y,z')

f = 3*x+4*y-2*z
g = 2*x-y+3*z
h = x+z

eq1 = Eq(f,3)
eq2 = Eq(g,2)
eq3 = Eq(h,4)

soluciones = solve((eq1,eq2,eq3),(x,y,z))
print(soluciones)
```
